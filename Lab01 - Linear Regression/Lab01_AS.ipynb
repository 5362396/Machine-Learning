# Regresja liniowa 

Import wszystkich wymaganych bibliotek:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

Import zestawu danych CSV 

```python
dataset = pd.read_csv('Kruk_VLagun_156.csv')
```

Sprawdzenie liczby wierszy i kolumn

```python
dataset.shape
```

Szczegóły statystyczne zbioru danych

```python
dataset.describe()
```

Wykres 2D między SS i SechhDisc

```python
dataset.plot(x='SS', y='SecchDisc', style='o')
plt.title('SS vs SecchDisc')
plt.xlabel('SS')
plt.ylabel('SecchDisc')
plt.show()
```

Wykres rozkładu widzialności w wodzie, aby zobaczyć gdzie możemy szukać jej średniej.

```python
plt.figure(figsize=(15, 10))
plt.tight_layout()
sns.distplot(dataset['SecchDisc'])
```

### Przewidywanie na podstawie regresji liniowej 

Podzielenie danych na "atrybuty" i "etykiety".

```python
X = dataset['SS'].values.reshape(-1, 1)
y = dataset['SecchDisc'].values.reshape(-1, 1)
```

Dzielenie 80% danych na zestaw treningowy, a 20% danych na zestaw testowy.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Trenowanie algorytmu.

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Wartość przechwycenia (intercept) i nachylenia (regressor coeficient).

```python
print(regressor.intercept_)
print(regressor.coef_)
```

Przewidywanie danych testowych.

```python
y_pred = regressor.predict(X_test)
```

Porównanie rzeczywiste wartości wyjściowe dla X_test z przewidywanymi wartościami.

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

Wizualizacja wyniku porównania jako wykres słupkowy.

```python
df1 = df.head(25)
df1.plot(kind='bar', figsize=(16, 10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

Wykreślenie linii prostej z danymi testowymi:

```python
plt.scatter(X_test, y_test, color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

### Miary błędów modelu regresji liniowej

Średni błąd bezwzględny (MAE)

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
```

Średni błąd kwadratowy (MSE)

```python
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
```

Średni błąd do kwadratu (RMSE)

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
